### Recap : evalauting classifiers

**Accuracy** - how many of our predictions are correct

**Precision** - how many of the positives were actually positives (TP/(TP+FP))

**Recall** - how many of the positives we'Re catching (TP/(TP+FN))

**F1** - takes precision and recall into one ratio that goes from 0-1

**Confusion Matrix** - to look at the raw numbers of TP, FP, TN, FN

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
def print_classifier_evluation(y_true, y_pred, model):
    print(f'Results for your model {model}:')
    print(f'Accuracy: {round(accuracy_score(y_true, y_pred),3)}')
    print(f'Precision: {round(precision_score(y_true, y_pred),3)}')
    print(f'Recall: {round(recall_score(y_true, y_pred),3)}')
    print(f'F1: {round(f1_score(y_true, y_pred),3)}')
    
    cm=confusion_matrix(y_true, y_pred,)
    print(cm)
    sns.heatmap(cm, annot=True, fmt = 'd', xticklabels= ['non-fraud','fraud'], yticklabels= ['non-fraud','fraud'])
```

```python
y_pred = dummy.predict(X_test)
```

```python
print_classifier_evluation(y_test, y_pred, DummyClassifier)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
print_classifier_evluation(y_test, y_pred, 'RandomForest')
```
